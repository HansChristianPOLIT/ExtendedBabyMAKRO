```python
model.find_ss(do_print=True)
```

```python
model.calc_jac(do_print=True)
```

# Find IRF

**Set Exogenous variables to steady state:**

```python
model.set_exo_ss()
```

**Introduce shock:**

```python
Gshock = 50
shock = 0.01*ss.G
persistence = 0.80**np.arange(Gshock)
sol.G[:Gshock] = ss.G + shock*persistence
```

```python
model.find_IRF()
```

```python
# for row 1
Models1 = [model, model1, model2]     
Sigma_F = [4.0, 5.0, 6.0]
for i in range(len(Models1)):
    Models1[i].par.sigma_F = Sigma_F[i]
    Models1[i].find_ss()
    Models1[i].calc_jac()
    Gshock = 50
    shock_G = 0.01*Models1[i].ss.G
    persistence_G = 0.80**np.arange(Gshock)
    Models1[i].sol.G[:Gshock] = Models1[i].ss.G + shock_G*persistence_G
    Models1[i].find_IRF()

# for row 2
Models2 = [model3, model4, model5]
Sigma_m = [1.01, 1.50, 1.75] 
for i in range(len(Models2)):
    Models2[i].par.sigma_m = Sigma_m[i]
    Models2[i].find_ss()
    Models2[i].calc_jac()
    Gshock = 50
    shock_G = 0.01*Models2[i].ss.G
    persistence_G = 0.80**np.arange(Gshock)
    Models2[i].sol.G[:Gshock] = Models2[i].ss.G + shock_G*persistence_G
    Models2[i].find_IRF()
    
# for row 3
Models3 = [model6, model7, model8]
Sigma_Y = [0.75, 1.01, 1.25] 
for i in range(len(Models3)):
    Models3[i].par.sigma_Y = Sigma_Y[i]
    Models3[i].find_ss()
    Models3[i].calc_jac()
    Gshock = 50
    shock_G = 0.01*Models3[i].ss.G
    persistence_G = 0.80**np.arange(Gshock)
    Models3[i].sol.G[:Gshock] = Models3[i].ss.G + shock_G*persistence_G
    Models3[i].find_IRF()
    
# for row 4
Models4 = [model9, model10, model11] 
Lambda_B = [0.2, 0.5, 0.8] 
for i in range(len(Models4)):
    Models4[i].par.lambda_B = Lambda_B[i]
    Models4[i].find_ss()
    Models4[i].calc_jac()
    Gshock = 50
    shock_G = 0.01*Models4[i].ss.G
    persistence_G = 0.80**np.arange(Gshock)
    Models4[i].sol.G[:Gshock] = Models4[i].ss.G + shock_G*persistence_G
    Models4[i].find_IRF()

# for row 5
Models5 = [model12, model13, model14]
Psi_0 = [3.0, 4.0, 15.0] 
for i in range(len(Models5)):
    Models5[i].par.Psi_0 = Psi_0[i]
    Models5[i].find_ss()
    Models5[i].calc_jac()
    Gshock = 50
    shock_G = 0.01*Models5[i].ss.G
    persistence_G = 0.80**np.arange(Gshock)
    Models5[i].sol.G[:Gshock] = Models5[i].ss.G + shock_G*persistence_G
    Models5[i].find_IRF()
```

```python
model = BabyMAKROModelClass(name='Extended') # create an instance of the model
par = model.par
ss = model.ss
sol = model.sol
model1 = model.copy()
model2 = model.copy()
```

```python
Models = [model, model1, model2]
Lambda = [0.05, 0.2, 0.5] 
    
for i in range(len(Models)):
    Models[i].par.Lambda = Lambda[i]
    Models[i].find_ss()
    Models[i].calc_jac()
    Gshock = 50
    shock_G = 0.01*Models[i].ss.G
    persistence_G = 0.80**np.arange(Gshock)
    Models[i].sol.G[:Gshock] = Models[i].ss.G + shock_G*persistence_G
    Models[i].find_IRF()
```
