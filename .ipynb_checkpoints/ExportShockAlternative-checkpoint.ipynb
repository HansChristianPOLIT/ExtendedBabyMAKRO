# Foreign Demand Shock: Alternative Shock - A Comparison

```python
%load_ext autoreload
%autoreload 2

import time
import pickle
import numpy as np
from scipy import optimize

import matplotlib.pyplot as plt   
plt.style.use('seaborn-whitegrid')
prop_cycle = plt.rcParams['axes.prop_cycle']
colors = prop_cycle.by_key()['color']

from BabyMAKROModel import BabyMAKROModelClass
```

# A Comparison

```python
model = BabyMAKROModelClass(name='Extended') # create an instance of the model
par = model.par
ss = model.ss
sol = model.sol
model1 = model.copy()
```

```python
model.find_ss()
model.calc_jac()
Tshock = 50
shock_chi = 0.01*model.ss.chi
persistence_chi = 0.8**np.arange(Tshock)
model.sol.chi[:Tshock] = model.ss.chi + shock_chi*persistence_chi
shock_p_f = 0.03*model.ss.P_F
persistence_p_f = 0.80**np.arange(Tshock)
model.sol.P_F[:Tshock] = model.ss.P_F + shock_p_f*persistence_p_f
model.find_IRF()
```

**Introduce Alternative Shock**

```python
model1.find_ss()
model1.calc_jac()
Tshock = 50
shock_chi = 0.01*model1.ss.chi
persistence_chi = 0.8**np.arange(Tshock)
model1.sol.chi[:Tshock] = model1.ss.chi + shock_chi*persistence_chi
shock_p_f = 0.03*model1.ss.P_F
persistence_p_f = 0.80**np.arange(Tshock)
model1.sol.P_F[:Tshock] = model1.ss.P_F + shock_p_f*persistence_p_f
shock_p_g = 0.03*ss.P_M_G
model1.sol.P_M_G[:Tshock] = ss.P_M_G + shock_p_g*persistence_p_f
shock_p_c = 0.03*ss.P_M_C
model1.sol.P_M_C[:Tshock] = ss.P_M_C + shock_p_c*persistence_p_f
shock_p_i = 0.03*ss.P_M_I
model1.sol.P_M_I[:Tshock] = ss.P_M_I + shock_p_i*persistence_p_f
shock_p_x = 0.03*ss.P_M_X
model1.sol.P_M_X[:Tshock] = ss.P_M_X + shock_p_x*persistence_p_f
model1.find_IRF()
```

# Impulse Response Figures

```python
T_IRF = 150

# a. iniate figure
fig = plt.figure(figsize=(2*6,1*6/1.5))

# b. create subplots
ax = fig.add_subplot(1,2,1)
ax.plot((model.sol.Y[:T_IRF]/model.ss.Y-1)*100,label=f'Baseline Shock', color = 'k', linestyle = '--')
ax.plot((model1.sol.Y[:T_IRF]/model1.ss.Y-1)*100,label=f'Alternative Shock')
ax.set_ylabel('% of s.s.')
ax.set_xlabel('Year')
ax.set_title('Production ($Y_t$)')
ax.legend(frameon=True)

ax = fig.add_subplot(1,2,2)
ax.plot((model.sol.M[:T_IRF]/model.ss.M-1)*100,label=f'Baseline Shock', color = 'k', linestyle = '--')
ax.plot((model1.sol.M[:T_IRF]/model1.ss.M-1)*100,label=f'Alternative Shock')
ax.set_ylabel('% of s.s.')
ax.set_xlabel('Year')
ax.set_title('Import ($M_{t}$)')
ax.legend(frameon=True)

# c. layout
fig.tight_layout(pad=1); # spacing between subplots
plt.savefig("ExportShock/AlternativeShock.pdf", format="pdf", bbox_inches="tight")
```
